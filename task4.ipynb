```python
Valid_plans=[199, 299, 399, 599]
def validcheck(recharge_amt):
    if recharge_amt<50:
        return False
    if recharge_amt in Valid_plans:
        return True
    else:
        return False
while True:
    recharge_amt=int(input("Enter the amount to recharge:"))
    if validcheck(recharge_amt):
        print("Recharge successful")
        break;
    else:
        print("Recharge not successful")
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12


    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)

    return fare

while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        if distance < 0:
            print("❌ Distance cannot be negative. Try again.\n")
            continue

        total_fare = calculate_fare(distance, peak_hour)


        print("✅ Total Cab Fare = ₹", total_fare)
        break

    except:
        print("❌ Invalid input. Please try again.\n")

```

```python
def check_eligibility(attendance):

    total_days = len(attendance)
    present_days = attendance.count("P")


    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = input("Enter attendance (P/A separated by space): ").split()

result = check_eligibility(attendance_list)
print(result)
```

```python
def check_password(password):
    if len(password) < 8:
        return "❌ Weak Password (Minimum 8 characters required)"

    has_digit = any(char.isdigit() for char in password)
    special_chars = "@#$"
    has_special = any(char in special_chars for char in password)

    if has_digit and has_special:
        return "✅ Strong Password"
    else:
        return "❌ Weak Password (Need digit and special character)"

user_password = input("Enter your password: ")
result = check_password(user_password)
print(result)
```
