```python
def analyze_engagement(likes):
    total_likes = 0

    for like in likes:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

post_likes = [120, 340, 210, 180, 200]
analyze_engagement(post_likes)
```

```python
def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock = 6
check_medicine_stock(medicine_stock)
```

```python
def analyze_rainfall(data, required_level):
    total = 0

    for value in data:
        total += value

    average = total // len(data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_data = [65, 70, 80, 75, 70]
analyze_rainfall(rainfall_data, 70)
```

```python
def detect_duplicates(usernames):
    unique_users = set()
    duplicate_found = False

    for name in usernames:
        if name in unique_users:
            duplicate_found = True
            break
        unique_users.add(name)

    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["alex", "john", "alex", "maria"]
detect_duplicates(users)
```

```python
def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

patient_age = 21
check_appointment_eligibility(patient_age)
```

```python
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

crop_prices = [1800, 2500, 1500, 3200, 1900]
filter_premium_crops(crop_prices)
```

```python
def evaluate_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

error_count = 7
evaluate_system_health(error_count)
```

```python
def validate_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_amount = 60000
validate_transaction(transaction_amount)
```

```python
def check_exam_eligibility(attendance):
    total_days = len(attendance)
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_list = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance_list)
```

```python
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


total_bill, status = calculate_electricity_bill(260)
print("Total Bill:", total_bill)
print("Usage Status:", status)
```
