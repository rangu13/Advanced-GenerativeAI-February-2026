```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def control(self, action):
        if action == "ON":
            self.status = "ON"
        else:
            self.status = "OFF"

    def show(self):
        print(self.name + " is " + self.status)

light_name = input()
action = input()

light = SmartLight(light_name)
light.control(action)
light.show()
```

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)

name = input()
emp_id = input()
dept = input()

emp = Employee(name, emp_id, dept)
emp.display()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

name = input()
phone = input()

c = Contact(name, phone)
c.show()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

name = input()
price = input()

p = Product(name, price)
p.display()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

name = input()
rating = input()

m = Movie(name, rating)
m.display()
```

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)

customer = input()
address = input()

d = Delivery(customer, address)
d.show()
```
