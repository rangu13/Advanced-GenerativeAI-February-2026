```python
sentence=input("Enter Sentence:").split()
count=0
unique_words=[]
for i in sentence:
    if i not in unique_words:
        count+=1
        unique_words.append(i)
print("Unique words count:",count)
print("Unique words:",unique_words)
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highemp=""
highsalary=0
for name,salary in employees.items():
    if salary > highsalary:
        highsalary=salary
        highemp=name
print("Highest Salary:",highemp,"-",highsalary)
```

```python
input_numbers=input("Enter the numbers by leaving one space aftet other:").split()
numbers=[]
for i in input_numbers:
    numbers.append(int(i))
maximum=numbers[0]
minimum=numbers[0]

for j in numbers:
    if j > maximum:
        maximum=j
    if j < minimum:
        minimum=j
print("Maximum:",maximum)
print("Minimum:",minimum)
```

```python
attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
    if i == "P":
        count += 1
percentage=count/len(attendance)*100
print("Attendance percentage:",percentage)
```
