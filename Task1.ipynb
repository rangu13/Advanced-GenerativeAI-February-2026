```python
marks = [45, 78, 90, 33, 60]
ps=0
fs=0
for i in marks:
    if i>=50:
        ps=ps+1
    else:
        fs=fs+1
print("Total number of student passed=",ps)
print("Total number of student failed=",fs)
```
